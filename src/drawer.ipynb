```python
import pickle
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
with open("../stats/fasion_mnist_Unet_linear.pkl", "rb") as f:
    linear_losses = pickle.load(f)
with open("../stats/fasion_mnist_Unet_cosine.pkl", "rb") as f:
    cosine_losses = pickle.load(f)
with open("../stats/fasion_mnist_DEQ_anderson_24.pkl", "rb") as f:
    deq_12 = pickle.load(f)
```
